Question 1: What is Dimensionality Reduction? Why is it important in machine learning?
Answer:

Dimensionality Reduction is the process of reducing the number of input features in a dataset while preserving as much important information as possible.
It transforms high-dimensional data into a lower-dimensional representation.

Importance in Machine Learning:

Reduces computational cost – fewer features → faster training.

Improves model performance by removing noise and irrelevant features.

Helps visualization of high-dimensional data in 2D or 3D.

Avoids overfitting by simplifying the model.


Question 2: Name and briefly describe three common dimensionality reduction techniques.
Answer:
1. Principal Component Analysis (PCA)

A linear technique that transforms data into new uncorrelated features (principal components) that capture maximum variance.

2. t-Distributed Stochastic Neighbor Embedding (t-SNE)

A nonlinear technique mainly used for visualizing high-dimensional data in 2D or 3D by preserving local structure.

3. Linear Discriminant Analysis (LDA)

A supervised method that reduces dimensions by maximizing class separability.


Question 3: What is clustering in unsupervised learning? Mention three popular clustering algorithms.
Answer:

Clustering is an unsupervised learning technique used to group similar data points together based on patterns or similarities without using labels.

Popular Clustering Algorithms:

K-Means Clustering – partitions data into K clusters.

Hierarchical Clustering – builds a tree (dendrogram) of clusters.

DBSCAN – density-based clustering that identifies core points and noise.


Question 4: Explain the concept of anomaly detection and its significance.
Answer:

Anomaly Detection refers to identifying rare, unusual, or abnormal data points that deviate from the majority of the data.

Significance:

Detects fraud, network intrusions, machine failures, or health abnormalities.

Ensures system reliability, security, and early warning for critical events.

Helps improve data quality and business decision-making.



Question 5: List and briefly describe three types of anomaly detection techniques.
Answer:
1. Statistical Methods

Use statistical thresholds (mean, standard deviation, probability distributions) to detect data points that are statistically unusual.

2. Machine Learning–Based Methods

Algorithms like Isolation Forest, One-Class SVM learn patterns of normal data and detect deviations.

3. Distance/Clustering-Based Methods

Use distance metrics (e.g., K-Means, DBSCAN) to detect points far from clusters or having low density.


Question 6: What is time series analysis? Mention two key components of time series data.
Answer:

Time Series Analysis involves studying data points collected at regular time intervals to identify trends, patterns, or seasonal behavior.

Key Components:

Trend – long-term upward or downward movement.

Seasonality – repeating patterns at regular intervals (daily, monthly, yearly).


Question 7: Describe the difference between seasonality and cyclic behavior in time series.
Answer:
Seasonality

Occurs at fixed, regular intervals such as daily, weekly, or yearly patterns.

Example: Increased ice-cream sales every summer.

Cyclic Behavior

Does not follow a fixed period; cycles are influenced by economic or natural processes.

Example: Economic recession cycles.



Question 8: Write Python code to perform K-means clustering on a sample dataset.
Answer (Copy-Paste Ready for Google Colab):
# K-Means Clustering Example

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate sample dataset
X, y = make_blobs(n_samples=300, centers=4, random_state=42)

# Apply K-Means
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)

# Cluster labels and centers
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# Plot the clustering result
plt.figure(figsize=(8,5))
plt.scatter(X[:,0], X[:,1], c=labels)
plt.scatter(centers[:,0], centers[:,1], s=200, marker='X')
plt.title("K-Means Clustering Output")
plt.show()





Question 9: What is inheritance in OOP? Provide a simple example in Python.
Answer:

Inheritance is an Object-Oriented Programming (OOP) concept where a new class (child) acquires attributes and methods of an existing class (parent).
It promotes code reusability and hierarchical structure.

Python Example:
# Example of Inheritance

class Animal:
    def sound(self):
        return "Some sound"

class Dog(Animal):  # Dog inherits from Animal
    def sound(self):
        return "Bark"

# Using the classes
a = Animal()
d = Dog()

print(a.sound())  # Output: Some sound
print(d.sound())  # Output: Bark






Question 10: How can time series analysis be used for anomaly detection?
Answer:

Time series analysis detects anomalies by identifying unusual patterns or deviations in temporal data.

Methods Used:

Forecasting models (ARIMA, LSTM) detect anomalies when actual values differ significantly from forecasted values.

Seasonality decomposition helps detect points that deviate from expected seasonal patterns.

Rolling statistics (moving average, moving variance) highlight sudden spikes or drops.

Change-point detection identifies structural breaks in data.

This helps detect equipment failures, fraud, sudden demand changes, or sensor faults.
